What are 2 main reason to sort a collection?
-----

1) Speed up search. Search on a sorted collection is 0(log n).



2) Identify similar items. Similar items will be grouped together after sorting.

When would you NOT sort before searching?
-----

1. When only searching a couple of times

Sorting costs O(n log n) and search on a sorted collection costs O(log n).

Just search on a collection is O(n).

For each data size and number of searches, there is a trade-off point. __Engineering is always about trade-offs__

2) Pervsere order of orginal data

If the order of the data matter (e.g., sensor data), then you might not want to sort.
